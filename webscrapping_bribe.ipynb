```python
import requests
from bs4 import BeautifulSoup
import urllib
import urllib.request 
import csv
import pandas as pd
import os
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=range(100)
pages=x
with open('C:/Users/Dell/Desktop/sanyoja/python/assignment/ipaidabribe50.csv','a',encoding='utf-8',newline='')as f_output:
    csv_print=csv.writer(f_output)
    
    file_is_empty=os.stat('C:/Users/Dell/Desktop/sanyoja/python/assignment/ipaidabribe50.csv').st_size==0
    if file_is_empty:
        csv_print.writerow(['bribe_title','name','transaction','amount','date','location','views'])

    for page in pages:
        
        url=requests.get("http://www.ipaidabribe.com/reports/all?page={}".format(page)).text 
        
        soup = BeautifulSoup(url, 'lxml')
        for bribe in soup.find_all("section",{"class":"ref-module-paid-bribe"}):
            
            try:
                bribe_title=(bribe.find("h3",{"class":"heading-3"}).text.strip())
            except Exception as e:
                bribe_title=None
            print('bribe_title:',bribe_title)

            try:
                name=(bribe.find("li",class_="name").text.strip())
            except Exception as e:
                name=None
            print('name:',name)

            try:
                transaction=(bribe.find("li",{"class":"transaction"}).text.strip())
            except Exception as e:
                transaction=None
            print('transaction:',transaction)

            try:
                 amount=(bribe.find("li",{"class":"paid-amount"}).text.strip())
            except Exception as e:
                 amount=None
            print('amount:',amount)

            try:
                date=(bribe.find("span",{"class":"date"}).text.strip())
            except Exception as e:
                date=None
            print('date:',date)

            try:
                location=(bribe.find("a",{"class":"location"}).text.strip())
            except Exception as e:
                location=None
            print('location:',location)

            try:
                views=(bribe.find("li",{"class":"views"}).text.strip())
            except Exception as e:
                views=None
            print('views:',views)

            csv_print.writerow([bribe_title,name,transaction,amount,date,location,views])

            print("-----------------------------------------------------------------------")


```

```python

df.amount = df.amount.str.strip().str.lower().str.replace("\\r\\n"," ").str.replace("paid inr", '').str.replace(",", "")
df.head()
```

```python
df.location = df.location.str.strip().str.lower().str.replace(' ', ' ').str.replace('\r\n                      ,', '').str.replace(')', '')
df.head()
```

```python
new = df["location"].str.split(" ", n = 1, expand = True) 
  
# making separate city column from new data frame 
df["city"]= new[0] 
  
# making separate state column from new data frame 
df["state"]= new[1] 
  
# Dropping old Name columns 
df.drop(columns =["location"], inplace = True) 
  
# df display 
df
```

```python
new = df["views"].str.split(" ", n = 1, expand = True) 
  
# making separate city column from new data frame 
df["views_no"]= new[0] 
  
# making separate state column from new data frame 
df["views_"]= new[1] 
  
# Dropping old Name columns 
df.drop(columns =["views"], inplace = True) 
  
# df display 
df.drop(columns =["views_"], inplace = True)
df
```
